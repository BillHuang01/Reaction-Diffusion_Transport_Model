```python
# import packages
import time
import numpy as np
import matplotlib.pyplot as plt
from TMA_PMMA import TMA_PMMA
```

```python
# please change only this part of the code
# define model parameters
l = 4.830e-5 # polymer thickness (cm)
df = 1.650e-10 # initial diffusivity (cm^2/s) 
sc = 4.436e-3 # surface concentraion (mol/cm^3)
pc = 5.656e-3 # polymer concentraion (mol/cm^3)
hd = 1150 # hindering factor (cm^3/mol)
k = 1 # reaction rate (cm^3/mol*s)
```

```python
# time discretization
times = np.concatenate((np.linspace(0,5000,2001),np.linspace(5000,125000,4801)[1:]))
```

```python
# initialize model and states
model_params = {'l':l, 'df':df, 'sc':sc, 'pc':pc, 'hd':hd, 'k':k}
model = TMA_PMMA(model_params)
s0 = model.init_state()
```

```python
# faster pde solver by using larger tolerance
start_time = time.time()
C, S, B = model.integrate(s0, times, rtol=1e-3, atol=1e-3)
end_time = time.time()
print('execution completes in %.2f seconds' %(end_time-start_time))
```

```python
# visualize mass uptake over time
fig = plt.figure(facecolor='w')
ax = fig.add_subplot(111, facecolor='#dddddd', axisbelow=True)
ax.plot(np.sqrt(times), C+S, lw=1)
ax.set_xlabel('time^0.5 (s^0.5)')
ax.set_ylabel('Mass/Area (ng/cm^2)')
plt.show()
```

```python
np.savetxt("mass_uptake.csv", np.transpose(np.vstack((times,C,S,B))), header="times,C,S,B", delimiter=',', comments="")
```

```python
# more accurate pde solver with default tolerance
start_time = time.time()
C, S, B = model.integrate(s0, times)
end_time = time.time()
print('execution completes in %.2f seconds' %(end_time-start_time))
```
